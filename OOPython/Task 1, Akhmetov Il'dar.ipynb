#### Class Citizen

```python
class Citizen:
    
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None):
        self.SetName(name)
        self.SetSurname(surname)
        self.SetAge(age)
        self.SetNationality(nationality)
        self.SetAddress(address)
    
    # setters
    def SetName(self, name):
        self._name = name
    
    def SetSurname(self, surname):
        self._surname = surname
    
    def SetAge(self, age): 
        self._age = age
    
    def SetNationality(self, nationality):
        self._nationality = nationality
    
    def SetAddress(self, address): 
        self._address = address
        
    # getters
    def GetName(self):
        return self._name
    
    def GetSurname(self):
        return self._surname
    
    def GetAge(self):
        return self._age
    
    def GetNationality(self):
        return self._nationality
    
    def GetAddress(self):
        return self._address
    
    def get_object_name(self):
        for k, v in globals().items():
            if v is self:
                return k
```

#### Class Worker

```python
class Worker(Citizen):
    
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None, \
                 companyName=None, companyAddress=None, workPhone=None):
        super().__init__(name, surname, age, nationality, address)
        self.SetCompanyName(companyName)
        self.SetCompanyAddress(companyAddress)
        self.SetWorkPhone(workPhone)
    
    #setters
    def SetCompanyName(self, companyName):
        self._companyName = companyName
        
    def SetCompanyAddress(self, companyAddress):
        self._companyAddress = companyAddress
        
    def SetWorkPhone(self, workPhone):
        self._workPhone = workPhone
    
    #getters
    def GetCompanyName(self):
        return self._companyName
    
    def GetCompanyAddress(self):
        return self._companyAddress
    
    def GetWorkPhone(self):
        return self._workPhone
```

#### Class Scientist

```python
from enum import Enum
class ScientificField(Enum):
    UNDEFINED = None
    MATH = 1
    PHYSICS = 2
    CHEMISTRY = 3

class ScientistType(Enum):
    UNDEFINED = None
    THEORIST = 1
    EXPERIMENTER = 2
    CALCULATOR = 3

class Scientist(Worker):
    
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None, \
                 companyName=None, companyAddress=None, workPhone=None, \
                 scientificField=None, scientificType=None, numberOfPublications=None):
        super().__init__(name, surname, age, nationality, address, companyName, companyAddress, workPhone)
        self.SetScientificField(scientificField)
        self.SetScientificType(scientificType)
        self.SetNumberOfPublications(numberOfPublications)

    #setters
    def SetScientificField(self, scientificField):
        self._scientificField = ScientificField(scientificField)
        
    def SetScientificType(self, scientificType):
        self._scientificType = ScientistType(scientificType)
        
    def SetNumberOfPublications(self, numberOfPublications):
        self._numberOfPublications = numberOfPublications
        
    #getters
    def GetScientificField(self):
        return self._scientificField.name
        
    def GetScientificType(self):
        return self._scientificType.name
        
    def GetNumberOfPublications(self):
        return self._numberOfPublications
```

#### Classes Master, Candidate, Doctor

```python
class Master(Scientist):
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None, \
                 companyName=None, companyAddress=None, workPhone=None, \
                 scientificField=None, scientificType=None, numberOfPublications=None):
        super().__init__(name, surname, age, nationality, address, \
                         companyName, companyAddress, workPhone, \
                         scientificField, scientificType, numberOfPublications)
class Candidate(Scientist):
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None, \
                 companyName=None, companyAddress=None, workPhone=None, \
                 scientificField=None, scientificType=None, numberOfPublications=None):
        super().__init__(name, surname, age, nationality, address, \
                         companyName, companyAddress, workPhone, \
                         scientificField, scientificType, numberOfPublications)
class Doctor(Scientist):
    def __init__(self, name=None, surname=None, age=None, nationality=None, address=None, \
                 companyName=None, companyAddress=None, workPhone=None, \
                 scientificField=None, scientificType=None, numberOfPublications=None):
        super().__init__(name, surname, age, nationality, address, \
                         companyName, companyAddress, workPhone, \
                         scientificField, scientificType, numberOfPublications)
```
